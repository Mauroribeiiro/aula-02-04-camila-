```python
#feito com ajuda
 
#decimo primeiro

#armazenar as temperaturas
temperaturas = []

# Solicita as 5 temperaturas
for i in range(5):
    temperatura = float(input(f"Digite a temperatura {i+1} em ºC: "))
    temperaturas.append(temperatura)

# Calcular média de temperatura 
media = sum(temperaturas) / 5

maior_temperatura = max(temperaturas)
menor_temperatura = min(temperaturas)

#resultados
print('Média das temperaturas:', (media), 'ºC')
print('Maior temperatura:', (maior_temperatura) ,'ºC')
print('Menor temperatura:', (menor_temperatura) ,'ºC')
```

```python
#décimo


# Solicitar o valor da hora e horas trabalhadas 
valor_hora = float(input("Digite o valor da sua hora de trabalho: "))
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas neste mês: "))

# Calcular o salário bruto
salario_bruto = valor_hora * horas_trabalhadas

# Calcular os descontos
imposto_renda = salario_bruto  # Defina a taxa de imposto de renda conforme a tabela
sindicato = salario_bruto * 0.03
fgts = salario_bruto * 0.11

#salário líquido
salario_liquido = salario_bruto - (imposto_renda + sindicato)


# Exibe os resultados
print('Salário Bruto: R$' (salario_bruto))
print('Desconto Imposto de Renda: R$',(imposto_renda))
print('Desconto Sindicato: R$',(sindicato))
print('FGTS (não descontado): R$', (fgts))
print('Salário Líquido: R$', (salario_liquido))

```
